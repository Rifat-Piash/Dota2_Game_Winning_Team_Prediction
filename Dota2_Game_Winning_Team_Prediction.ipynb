```python
def findMissingValue(df):
    # Calculate the number of missing values per column
    missing_values = df.isnull().sum()

    # Calculate the percentage of missing values
    percent_missing = (missing_values / len(df)) * 100

    # Create a DataFrame with the results
    missing_data = pd.DataFrame({
        'Columns': df.columns,
        'Number_Of_MValue': missing_values,
        'Percentage_Of_MValue': percent_missing.round(2)
    })

    # Filter out columns with no missing data and sort
    missing_data = missing_data[missing_data['Number_Of_MValue'] > 0].sort_values('Number_Of_MValue', ascending=False).reset_index(drop=True)

    return missing_data
```

<h2><a class="anchor" id="Section_3_2"></a><font color=blue>3.2 Variable Creation</font></h2>
<br/><br/>
<p style="font-size:16px;line-height:2.5em;">
In this phase, new features were created to enhance the predictive power of the model. The primary focus was on calculating the win rates for each hero based on their performance in historical matches. The process began by initializing a dictionary to hold the wins and losses for each hero present in the dataset.
</p>
<p style="font-size:16px;line-height:2.5em;">
The win/loss data was extracted by iterating through each row of the training dataset. If a match resulted in a win for the team (indicated by a target value of +1), the heroes in that team (marked with +1) had their win counts incremented, while heroes in the opposing team (marked with -1) had their loss counts incremented. Conversely, if the target value was -1, indicating a win for the opposing team, the heroes marked with -1 had their win counts incremented, and those marked with +1 had their loss counts incremented.
</p>
<p style="font-size:16px;line-height:2.5em;">
After tallying wins and losses, the win percentage for each hero was calculated. This involved determining the total number of games played by each hero and computing the percentage of games won relative to their total participation. The win percentages were subsequently stored back in the dictionary for each hero.
</p>
<p style="font-size:16px;line-height:2.5em;">
To further enrich the dataset, two new features were created to capture the average winning and losing percentages for the heroes on each team. This was achieved by defining a function that calculated these averages for the heroes indicated in each match row. The resulting averages were then added as new columns to the original DataFrame, enhancing the model's ability to assess the overall team composition in terms of past performance.
</p>
<p style="font-size:16px;line-height:2.5em;">
Ultimately, these newly created features will be instrumental in predicting the outcomes of Dota 2 matches based on team compositions, providing valuable insights for players when forming their teams.
</p>
<br/><br/>
<a href="#tbc">Go Back To Table Of Content</a>


```python
# Initialize a dictionary to hold wins and losses for each hero
win_rates = {hero: {'wins': 0, 'losses': 0} for hero in df_train.columns[4:]}  # Assuming hero names start from index 4

# Calculate wins and losses for each hero based on the target column
for index, row in df_train.iterrows():
    if row['target'] == 1:  # Winning team (heroes with +1 won)
        # Increment win count for heroes that are in the winning team (value == 1)
        for hero in df_train.columns[4:]:  # Iterate through hero columns
            if row[hero] == 1:  # Check if the hero is on the winning team
                win_rates[hero]['wins'] += 1
            elif row[hero] == -1:  # Check if the hero is on the losing team
                win_rates[hero]['losses'] += 1  # Count as a loss for heroes in the losing team
    elif row['target'] == -1:  # Losing team (heroes with -1 won)
        # Increment win count for heroes that are in the winning team (value == -1)
        for hero in df_train.columns[4:]:  # Iterate through hero columns
            if row[hero] == -1:  # Check if the hero is on the winning team
                win_rates[hero]['wins'] += 1
            elif row[hero] == 1:  # Check if the hero is on the losing team
                win_rates[hero]['losses'] += 1  # Count as a loss for heroes in the losing team

# Calculate win percentages for each hero
for hero, data in win_rates.items():
    total_games = data['wins'] + data['losses']
    win_percentage = (data['wins'] / total_games * 100) if total_games > 0 else 0
    win_rates[hero]['win_percentage'] = win_percentage

# Create new features in the original DataFrame for team averages
def calculate_team_percentages(row):
    team_1_percentages = [win_rates[hero]['win_percentage'] for hero in df_train.columns[4:] if row[hero] == 1]
    team_2_percentages = [win_rates[hero]['win_percentage'] for hero in df_train.columns[4:] if row[hero] == -1]
    
    avg_win_rate_1 = sum(team_1_percentages) / len(team_1_percentages) if team_1_percentages else 0
    avg_win_rate_2 = sum(team_2_percentages) / len(team_2_percentages) if team_2_percentages else 0
    
    return avg_win_rate_1, avg_win_rate_2

# Apply the function to create new columns for average winning and losing percentages
df_train[['avg_win_rate_1', 'avg_win_rate_2']] = df_train.apply(
    calculate_team_percentages,
    axis=1, result_type='expand'
)


# Show the updated DataFrame
df_train.head()
```

<h2><a class="anchor" id="Section_4_2"></a><font color=blue>4.2 Building Machine Learning Models</font></h2>
<br/><br/>
<p style="font-size:16px;line-height:2.5em;">
Building machine learning models such as Random Forest, Decision Tree, Logistic Regression, and XGBoost. The performance of each model will be evaluated to track progress and assess improvements achieved before and after hyperparameter tuning.
</p>
<br/><br/>
<a href="#tbc">Go Back To Table Of Content</a>


```python
# Building and Evaluating Models with K-Fold Cross-Validation and Testing on Unseen Data

# List of models to evaluate
models = [
    ("Random Forest", RandomForestClassifier()),
    ("Decision Tree", DecisionTreeClassifier()),
    ("Logistic Regression", LogisticRegression(max_iter=1000)),
    ("XGBoost", XGBClassifier())
]

# Data structure to store all results
results = []
unseen_data_results = []

# Building model function
def building_model(models):
    for model_name, model in models:
        print(f"\nEvaluating Model: {model_name}")
        
        # Parameters
        k = 2  # Number of folds

        # Initialize Stratified K-Fold
        skf = StratifiedKFold(n_splits=k)

        # Lists to store performance metrics for the current model
        accuracy_list = []
        roc_auc_list = []
        precision_list = []
        recall_list = []
        f1_list = []

        # K-Fold Cross-Validation
        for fold, (train_index, test_index) in enumerate(skf.split(X_for_initial_model, y_for_initial_model)):
            print(f"\nFold {fold + 1}/{k}")

            # Split data into training and testing sets
            X_train, X_test = X_for_initial_model.iloc[train_index], X_for_initial_model.iloc[test_index]
            y_train, y_test = y_for_initial_model.iloc[train_index], y_for_initial_model.iloc[test_index]

            # BalancedBaggingClassifier with the current model as base estimator
            bbc_model = BalancedBaggingClassifier(
                base_estimator=model,
                sampling_strategy='auto',
                replacement=False,
                random_state=1
            )

            # Train the classifier
            bbc_model.fit(X_train, y_train)

            # Make predictions
            predictions = bbc_model.predict(X_test)

            # Evaluate performance metrics
            accuracy = bbc_model.score(X_test, y_test)
            roc_auc = roc_auc_score(y_test, predictions)
            report = classification_report(y_test, predictions, output_dict=True)

            precision = report['weighted avg']['precision']
            recall = report['weighted avg']['recall']
            f1 = report['weighted avg']['f1-score']

            # Append metrics for each fold
            accuracy_list.append(accuracy)
            roc_auc_list.append(roc_auc)
            precision_list.append(precision)
            recall_list.append(recall)
            f1_list.append(f1)

            # Print the metrics for each fold
            print(f"Accuracy: {accuracy:.4f}")
            print(f"ROC AUC: {roc_auc:.4f}")
            print(f"Precision: {precision:.4f}")
            print(f"Recall: {recall:.4f}")
            print(f"F1-Score: {f1:.4f}")

        # Calculate the average metrics for the current model
        avg_accuracy = np.mean(accuracy_list).round(4)
        avg_roc_auc = np.mean(roc_auc_list).round(4)
        avg_precision = np.mean(precision_list).round(4)
        avg_recall = np.mean(recall_list).round(4)
        avg_f1 = np.mean(f1_list).round(4)

        # Append results to the results list
        results.append({
            'Model': model_name,
            'Average Accuracy': avg_accuracy,
            'Average ROC AUC': avg_roc_auc,
            'Average Precision': avg_precision,
            'Average Recall': avg_recall,
            'Average F1-Score': avg_f1
        })

        # Test on unseen data

        test = df_test.copy()

        validate = test.drop(columns='target')

        # Save and load the model using pickle
        saved_model = pickle.dumps(bbc_model)
        model_from_pickle = pickle.loads(saved_model)

        # Make predictions on unseen data
        validate['Prediction'] = model_from_pickle.predict(validate)
        validate['target'] = test['target']
        val_result = validate[['Prediction', 'target']].copy()
        val_result.reset_index(drop=True, inplace=True)
        val_result['Total'] = val_result['Prediction'] + val_result['target']
        
        # Calculate accuracy on unseen data
        unseen_accuracy = (len(val_result[(val_result['Total'] ==2)|(val_result['Total'] ==-2)])/len(val_result))

        # Append unseen data accuracy to unseen_data_results list
        unseen_data_results.append({
            'Model': model_name,
            'Unseen Data Accuracy': unseen_accuracy
        })

        # Print the average metrics for the model
        print("\n=== K-Fold Cross-Validation Results ===")
        print(f"Model: {model_name}")
        print(f"Average Accuracy: {avg_accuracy}")
        print(f"Average ROC AUC: {avg_roc_auc}")
        print(f"Average Precision: {avg_precision}")
        print(f"Average Recall: {avg_recall}")
        print(f"Average F1-Score: {avg_f1}")
        print(f"Unseen Data Accuracy: {unseen_accuracy:.2f}%")

# Run the function with the list of models
building_model(models)

# Convert results to a DataFrame for summary
results_df = pd.DataFrame(results)
unseen_results_df = pd.DataFrame(unseen_data_results)
```

```python
# # Building XGBoost Model

# Parameters
k = 3  # Number of folds

# Initialize Stratified K-Fold
skf = StratifiedKFold(n_splits=k)

# Lists to store performance metrics
accuracy_list = []
roc_auc_list = []
precision_list = []
recall_list = []
f1_list = []

# K-Fold Cross-Validation
for fold, (train_index, test_index) in enumerate(skf.split(X, y)):
    
    print(f"Fold {fold + 1}/{k}")
    
    # Split data into training and testing sets
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]   

    
    # Step 3: BalancedBaggingClassifier with XGBoost as base estimator
    bbc_xg = BalancedBaggingClassifier(
        base_estimator=XGBClassifier(kernel='linear',gamma=0.1,C=0.1),
        sampling_strategy='auto',
        replacement=False,
        random_state=1
    )
    
    # Train the classifier
    bbc_xg.fit(X_train, y_train)
    
    # Make predictions
    predictions = bbc_xg.predict(X_test)
    
    # Evaluate performance metrics
    accuracy = bbc_xg.score(X_test, y_test)
    roc_auc = roc_auc_score(y_test, predictions)
    report = classification_report(y_test, predictions, output_dict=True)
    
    precision = report['weighted avg']['precision']
    recall = report['weighted avg']['recall']
    f1 = report['weighted avg']['f1-score']
    
    # Append metrics to list
    accuracy_list.append(accuracy)
    roc_auc_list.append(roc_auc)
    precision_list.append(precision)
    recall_list.append(recall)
    f1_list.append(f1)
    
    # Print the metrics for each fold
    print(f"Accuracy: {accuracy:.4f}")
    print(f"ROC AUC: {roc_auc:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")
    print("\n")

model_name.append('XGBoost')
model_accuracy.append(np.mean(accuracy_list).round(4))
model_roc_auc.append(np.mean(roc_auc_list).round(4))
model_precision.append(np.mean(precision_list).round(4))
model_recall.append(np.mean(recall_list).round(4))
model_f1_score.append(np.mean(f1_list).round(4))  
    
# After cross-validation, print average performance
print("=== K-Fold Cross-Validation Results ===")
print(f"Average Accuracy: {np.mean(accuracy_list):.4f}")
print(f"Average ROC AUC: {np.mean(roc_auc_list):.4f}")
print(f"Average Precision: {np.mean(precision_list):.4f}")
print(f"Average Recall: {np.mean(recall_list):.4f}")
print(f"Average F1-Score: {np.mean(f1_list):.4f}")
```
